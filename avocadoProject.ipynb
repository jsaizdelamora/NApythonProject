```python
#Determining when the best time to buy avocados to sell at markup was, based on lowest price of Avocados

import pandas as pd
avocados = pd.read_csv("avocado.csv")

#sort listings by lowest price
sortedAvocados = avocados.sort_values(by="AveragePrice")
#find the cheapestListing
cheapestAvocado = sortedAvocados.head(1)
#pull relative date and year of listing
relativeDate = cheapestAvocado["Date"]
relativeYear = cheapestAvocado.iloc[0]['year']

#find comparative high price of avocado for same year
comparatorAvocado = avocados.loc[avocados["year"]==relativeYear ]
comparatorAvocado.sort_values(by="AveragePrice", ascending = False)

print("Highest price differential for avocados in "+ str(relativeYear))
print("Lowest Avocado price was "+ str(cheapestAvocado.iloc[0]["AveragePrice"])+" in "+str(cheapestAvocado.iloc[0]["region"]) )
print("Highest Avocado price was "+str(comparatorAvocado.iloc[0]["AveragePrice"])+" in "+str(comparatorAvocado.iloc[0]["region"] ))
```
